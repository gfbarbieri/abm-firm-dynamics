```python
import pandas as pd
import random
import math
```

```python
def increment(w, mu, sigma, scale):
    W = random.normalvariate(mu=0, sigma=1)
    inc_w = w * math.exp((mu - (sigma ** 2) / 2) + sigma * W)
    inc_w_x = w * math.exp((scale*mu - ((scale*sigma) ** 2) / 2) + (scale*sigma) * W)
    # inc_w_x = w * math.exp((scale*mu - ((sigma) ** 2) / 2) + (sigma) * W)

    return inc_w, inc_w_x
```

```python
# Inputs.
w = 1
mu = random.uniform(0, 1)
sigma = random.uniform(0, math.sqrt(2 * mu))

# Simulation
data = []

for i in range(1000):
    w, w_x = increment(w=w, mu=mu, sigma=sigma, scale=2)
    data.append([w, w_x, w_x / w])

# Plot Results
df = pd.DataFrame(data=data, columns=['w','w_x','r'])
df['r_m'] = df['r'].mean()
df[['r', 'r_m']].plot()
print(df['r'].mean())
```
